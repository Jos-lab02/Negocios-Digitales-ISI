### ¿Qué preguntas deberíamos plantearnos para obtener las respuestas que buscamos?

Para determinar cuál de las dos estrategias, M-Commerce o E-Commerce, es más rentable, es necesario plantear preguntas que permitan analizar el desempeño del negocio desde distintos ángulos. En primer lugar, se debe considerar qué canal genera mayores ingresos totales, ya que este indicador refleja el impacto directo en la rentabilidad. También es importante analizar qué canal presenta un mayor ticket promedio por transacción, debido a que permite identificar dónde los clientes gastan más en cada compra. Además, se debe evaluar cuál canal tiene mayor volumen de transacciones, ya que una mayor frecuencia de compra puede compensar un ticket promedio más bajo. Finalmente, es relevante observar cómo varía el comportamiento de ventas según la categoría de producto y el canal de venta, con el fin de entender qué tipo de productos se desempeñan mejor en cada estrategia.

En el análisis del dataset sintético se observa una distribución de uso de los canales donde aproximadamente el 60 por ciento de las transacciones corresponden a App Móvil (M-Commerce) y el 40 por ciento a Web Desktop (E-Commerce). Esta proporción sugiere una mayor adopción del canal móvil, lo cual podría influir directamente en el volumen de ventas y en la frecuencia de compra.


### ¿Por qué se crean 4 KPIs para responder esta pregunta?

Se definen cuatro KPIs dentro del sistema de información gerencial porque la rentabilidad no puede medirse únicamente con un solo indicador. La rentabilidad depende de múltiples factores que deben analizarse en conjunto, como los ingresos generados por cada canal, el valor promedio de las transacciones, el número de compras realizadas y el comportamiento de los clientes según el canal utilizado.

En este caso, los KPIs permiten transformar los datos operativos en información estratégica, facilitando la comparación entre M-Commerce y E-Commerce. Cada KPI aporta una visión diferente del desempeño, lo que permite construir un análisis más completo y confiable para la toma de decisiones.



### Conclusión

A partir del análisis realizado con los KPIs implementados en el dataset, se observa que la comparación entre M-Commerce y E-Commerce no depende de un solo indicador, sino del comportamiento conjunto de varias métricas.

El canal App Móvil presenta una mayor participación en el volumen de transacciones (aproximadamente 60 por ciento), lo cual sugiere una mayor adopción por parte de los usuarios. Sin embargo, la rentabilidad final debe evaluarse considerando simultáneamente el ingreso total, el ticket promedio y la frecuencia de compra.

Con base en los KPIs calculados, la decisión final indica que el canal con mejor desempeño global será aquel que logre mantener consistentemente mayor generación de ingresos junto con un equilibrio entre ticket promedio y volumen de transacciones. En este caso, el análisis permite identificar cuál de las dos estrategias es más eficiente en términos de rentabilidad operativa.

Por lo tanto, la decisión no se basa en un único indicador aislado, sino en la integración de todos los KPIs definidos, lo que permite una evaluación más robusta y orientada a la toma de decisiones estratégicas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Generar datos simulados
data = {
    'ID_Transaccion': range(1001, 1051),
    'Canal': np.random.choice(['Web Desktop', 'App Móvil'], 50, p=[0.4, 0.6]),
    'Metodo_Pago': np.random.choice(['Tarjeta Crédito', 'PayPal', 'Transferencia'], 50),
    'Total_Venta': np.random.randint(20, 500, 50),
    'Categoria': np.random.choice(['Electrónica', 'Moda', 'Hogar'], 50)
}

df = pd.DataFrame(data)

print(df)

# KPI 1: VENTAS TOTALES POR CANAL
ventas_canal = df.groupby('Canal')['Total_Venta'].sum()

print("\nKPI 1 - Ventas Totales por Canal")
print(ventas_canal)

plt.figure()
ventas_canal.plot(kind='bar')
plt.title('KPI 1 - Ventas Totales por Canal')
plt.ylabel('Ventas')
plt.show()


# KPI 2: TICKET PROMEDIO POR CANAL
ticket_canal = df.groupby('Canal')['Total_Venta'].mean()

print("\nKPI 2 - Ticket Promedio por Canal")
print(ticket_canal)

plt.figure()
ticket_canal.plot(kind='bar')
plt.title('KPI 2 - Ticket Promedio por Canal')
plt.ylabel('Promedio')
plt.show()


# KPI 3: TRANSACCIONES POR CANAL
transacciones = df.groupby('Canal')['ID_Transaccion'].count()

print("\nKPI 3 - Transacciones por Canal")
print(transacciones)

plt.figure()
transacciones.plot(kind='bar')
plt.title('KPI 3 - Transacciones por Canal')
plt.ylabel('Cantidad')
plt.show()


# KPI 4: RESUMEN DE RENTABILIDAD
resumen = pd.DataFrame({
    'Ventas Totales': ventas_canal,
    'Ticket Promedio': ticket_canal,
    'Transacciones': transacciones
})

print("\nKPI 4 - Resumen de Rentabilidad")
print(resumen)

plt.figure()
resumen.plot(kind='bar')
plt.title('KPI 4 - Comparación Final M-Commerce vs E-Commerce')
plt.show()


# CONCLUSIÓN PARTE 1
print("\nCONCLUSIÓN PARTE 1")

if ventas_canal['App Móvil'] > ventas_canal['Web Desktop']:
    mejor_canal = "App Móvil (M-Commerce)"
else:
    mejor_canal = "Web Desktop (E-Commerce)"

print("Canal más rentable:", mejor_canal)
print("Basado en ventas totales, ticket promedio y volumen de transacciones.")


# =====================================================
# PARTE 2: DATASET ECOMMERCE
# =====================================================

data = pd.read_csv('ecommerce.zip')


# KPI 1.1: COSTO PROMEDIO POR ENVÍO
kpi_1_1 = data.groupby('Mode_of_Shipment')['Cost_of_the_Product'].mean()

print("\nKPI 1.1 - Costo por modo de envío")
print(kpi_1_1)

plt.figure()
kpi_1_1.plot(kind='bar')
plt.title('KPI 1.1 - Costo por Modo de Envío')
plt.ylabel('Costo promedio')
plt.show()


# KPI 1.2: OCUPACIÓN DE BODEGA
kpi_1_2 = data.groupby('Warehouse_block')['Weight_in_gms'].sum()

print("\nKPI 1.2 - Ocupación de bodega")
print(kpi_1_2)

plt.figure()
kpi_1_2.plot(kind='bar')
plt.title('KPI 1.2 - Ocupación de Bodega')
plt.ylabel('Peso total')
plt.show()


# KPI 1.3: PESO PROMEDIO POR ENVÍO
kpi_1_3 = data.groupby('Mode_of_Shipment')['Weight_in_gms'].mean()

print("\nKPI 1.3 - Peso promedio por envío")
print(kpi_1_3)

plt.figure()
kpi_1_3.plot(kind='bar')
plt.title('KPI 1.3 - Peso Promedio por Envío')
plt.ylabel('Peso promedio')
plt.show()


# KPI 2.1: INGRESO Y COSTO
ingreso_total = data['Cost_of_the_Product'].sum()
costo_promedio = data['Cost_of_the_Product'].mean()

print("\nKPI 2.1")
print("Ingreso Total:", ingreso_total)
print("Costo Promedio:", costo_promedio)

#FIX: gráfico KPI 2.1
plt.figure()
plt.bar(['Ingreso Total', 'Costo Promedio'], [ingreso_total, costo_promedio])
plt.title('KPI 2.1 - Ingreso vs Costo')
plt.show()


# KPI 2.2: DESCUENTO PROMEDIO
descuento_promedio = data['Discount_offered'].mean()

print("\nKPI 2.2")
print("Descuento Promedio:", descuento_promedio)

#FIX: gráfico KPI 2.2
plt.figure()
plt.bar(['Descuento Promedio'], [descuento_promedio])
plt.title('KPI 2.2 - Descuento Promedio')
plt.show()


# KPI 2.3: IMPORTANCIA DEL PRODUCTO
kpi_2_3 = data.groupby('Product_importance')[['Cost_of_the_Product','Discount_offered']].mean()

print("\nKPI 2.3 - Importancia del producto")
print(kpi_2_3)

plt.figure()
kpi_2_3.plot(kind='bar')
plt.title('KPI 2.3 - Importancia vs Costos y Descuentos')
plt.show()


# KPI 3.1: CSAT
csat = data['Customer_rating'].mean()

print("\nKPI 3.1 - CSAT Promedio")
print(csat)

csat_genero = data.groupby('Gender')['Customer_rating'].mean()

plt.figure()
csat_genero.plot(kind='bar')
plt.title('KPI 3.1 - CSAT por Género')
plt.show()


# KPI 3.2: RETENCIÓN
nuevos = len(data[data['Prior_purchases'] <= 1])
recurrentes = len(data[data['Prior_purchases'] >= 2])

print("\nKPI 3.2 - Retención")
print("Clientes Nuevos %:", nuevos/len(data)*100)
print("Clientes Recurrentes %:", recurrentes/len(data)*100)

retencion = pd.Series({
    "Nuevos": nuevos,
    "Recurrentes": recurrentes
})

plt.figure()
retencion.plot(kind='bar')
plt.title('KPI 3.2 - Clientes Nuevos vs Recurrentes')
plt.show()


# KPI 3.3: SOPORTE
print("\nKPI 3.3 - Soporte")
print("Llamadas promedio:", data['Customer_care_calls'].mean())

llamadas_rating = data.groupby('Customer_rating')['Customer_care_calls'].mean()

plt.figure()
llamadas_rating.plot(marker='o')
plt.title('KPI 3.3 - Llamadas vs Calificación')
plt.xlabel('Rating')
plt.ylabel('Llamadas promedio')
plt.grid()
plt.show()
```
